## R part
We train Self-Organizing Map similarly to the original R example. 

```python
%%R
require(kohonen)
```

Following runs in R!

```python
%%R -i df -o X

# Rename data frame
data = df

# Get all exogenous variables, dropping endogenous ("admit")
# z-score normalization for exogenous variables
X = scale(data[,-1])

# Set seed for reproducibility
set.seed(222)

# Define SOM grid
g = somgrid(xdim = 4, ydim = 4, topo = "rectangular")

# Train SOM
map = som(X,grid=g,alpha=c(0.05,0.01), radius=1)

# Visualize what kind of observations each node contains on average
plot(map, type="codes")
```

```python
# We can inspect from R exported array in Python
X = pd.DataFrame(np.array(X))
display(X.head())
```
